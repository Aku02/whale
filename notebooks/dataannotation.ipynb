```python
import pandas as pd
import matplotlib.pyplot as plt
import cv2
import os
import numpy as np
import glob
import shutil
from tqdm.auto import tqdm

img_size_df = pd.read_csv('../input/whaleimgsize/train_image.csv')
img_size_df.head()
img_w_map = dict(zip(img_size_df['image'], img_size_df['w']))
img_size_df = img_size_df[img_size_df['w'] >= 3000].sample(6000)
img_size_df
```

```python

os.makedirs('data/train_images', exist_ok=True)

MAX_SIZE = 768

def copy_imgs(src, dest):
    apply_box = 0
    nan_box = 0
    bar = tqdm(img_size_df.iterrows(), total=len(img_size_df))
    for i, row in bar:
        img_file = os.path.join(src, row['image'])
        img_w = img_w_map[os.path.basename(img_file)]
        if img_w >= 3000:
            shutil.copy(img_file, f'{dest}/{os.path.basename(img_file)}')
            apply_box+=1
            #cv2.imwrite(f'{dest}/{os.path.basename(img_file)}')
            bar.set_description(f'copied {apply_box}')

    return apply_box, nan_box
            
print(copy_imgs('../input/happy-whale-and-dolphin/train_images', 'data/train_images'))
```
